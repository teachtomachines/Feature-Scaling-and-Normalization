# Feature Scaling and Normalization

```python
#use wine dataset

from sklearn.datasets import load_wine
import pandas as pd
%matplotlib inline

data = load_wine()
features = pd.DataFrame(data.data, columns=data.feature_names)

features
```

## MinMaxScaler

```python
#apply min-max scaler

from sklearn.preprocessing import MinMaxScaler

scaler   = MinMaxScaler()
scaler.fit(features)
scaled_features = scaler.transform(features)
scaled_features = pd.DataFrame(scaled_features, columns = features.keys())
```

```python
scaled_features
```

## Z Score Normalization

```python
#apply z score normalization

from sklearn.preprocessing import StandardScaler

normalizer = StandardScaler()
normalizer.fit(features)
features_z_scored = normalizer.transform(features)
features_z_scored = pd.DataFrame(features_z_scored, columns = features.keys())

features_z_scored
```

```python
#distribution of alcohol feature BEFORE normalization

features.alcohol.hist()
```

```python
#distribution of alcohol feature AFTER normalization

features_z_scored.alcohol.hist()
```
